```python
#to calculate euclidean dist between wrist_loc and sword_loc
import  math

def calc_euclidean(a, b):
    dist = (a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2    
    return math.sqrt(dist)
```
